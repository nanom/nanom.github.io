```haskell
separar :: [a]-> (a->Bool) -> ([a],[a])
separar [] _ = ([],[])
separar xs p = (iterar xs p, iterar xs (not.p))
    where   iterar [] _ = []
            iterar (x:xs) p
                | p x = x: iterar xs p
                | otherwise = iterar xs p
```

```haskell
separar [1..10] even
```
